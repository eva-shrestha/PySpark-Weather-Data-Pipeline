```
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```
spark = SparkSession.builder.getOrCreate()
```

```
df= spark.read.format("delta").load("dbfs:/user/hive/warehouse/cities")
df = df.select("*").limit(5)
df = df.withColumn('id', col('id').cast('int'))
display(df)
```

```
dimcity_df = df.select("id","lat","lon","name","country")
```

```
dimcity_df.show()
```

```
dimcity_df.write.format("delta").mode("overwrite").saveAsTable("dimcity")
```

```
%sql
select * from dimcity
```
